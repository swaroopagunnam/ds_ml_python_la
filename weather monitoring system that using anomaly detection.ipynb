```python
# First, install the required package
import sys
import subprocess
subprocess.check_call([sys.executable, '-m', 'pip', 'install', 'scikit-learn', 'numpy'])

# Now import the necessary libraries
import numpy as np
import logging
import time
from sklearn.ensemble import IsolationForest  # This is likely the anomaly detection model being used

# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Define global variables
anomaly_model = None

def read_sensor_data():
    # Mock function to simulate reading sensor data
    # In a real application, this would interface with actual sensors
    import random
    return {
        'temperature': round(random.uniform(15, 30), 1),
        'humidity': round(random.uniform(30, 90), 1),
        'pressure': round(random.uniform(990, 1020), 1)
    }

def notify_anomaly(data):
    # Mock function to handle anomaly notifications
    logging.warning(f"ANOMALY DETECTED: Temp={data['temperature']}°C, "
                   f"Hum={data['humidity']}%, "
                   f"Pres={data['pressure']}hPa, "
                   f"Score={data['anomaly_score']}")

def initialize_model():
    global anomaly_model
    # Initialize the anomaly detection model
    # In a real application, you might load a pre-trained model
    anomaly_model = IsolationForest(contamination=0.05)
    
    # Generate some normal data to train the model
    # This is just for demonstration purposes
    import random
    normal_data = []
    for _ in range(100):
        normal_data.append([
            random.uniform(15, 30),  # temperature
            random.uniform(30, 90),  # humidity
            random.uniform(990, 1020)  # pressure
        ])
    
    # Train the model
    anomaly_model.fit(normal_data)
    logging.info("Anomaly detection model initialized")

def run_weather_station():
    # Main function to run the weather station with anomaly analysis
    recent_readings = []
    max_readings = 100  # Store last 100 readings
    
    while True:
        try:
            sensor_data = read_sensor_data()
            
            if sensor_data and anomaly_model is not None:
                # Prepare data for the model (requires a 2D array)
                data_point = np.array([[
                    sensor_data['temperature'], 
                    sensor_data['humidity'], 
                    sensor_data['pressure']
                ]])
                
                # Predict if the new data point is an anomaly
                prediction = anomaly_model.predict(data_point)[0]
                anomaly_score = anomaly_model.decision_function(data_point)[0]
                
                # Add score to the data
                sensor_data['anomaly_score'] = round(anomaly_score, 4)
                
                # Store the reading
                recent_readings.append(sensor_data)
                if len(recent_readings) > max_readings:
                    recent_readings.pop(0)
                
                if prediction == -1:
                    notify_anomaly(sensor_data)
                else:
                    logging.info(f"Normal data: Temp={sensor_data['temperature']}°C, "
                                f"Hum={sensor_data['humidity']}%, "
                                f"Pres={sensor_data['pressure']}hPa, "
                                f"Score={sensor_data['anomaly_score']}")
            
            # Wait before the next reading
            time.sleep(5)
        except KeyboardInterrupt:
            logging.info("Weather station stopped by user.")
            break
        except Exception as e:
            logging.error(f"Error in main loop: {e}")
            time.sleep(5)  # Wait before retrying

# Entry point of the script
if __name__ == "__main__":
    try:
        initialize_model()
        run_weather_station()
    except Exception as e:
        logging.critical(f"Fatal error: {e}")
```
