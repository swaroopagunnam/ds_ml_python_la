```python
"""Task 5: Calculate Sum of a List"""
#develop a Python function called calculate_sum that takes a list of numbers as
#input and returns the sum of the numbers.
#(Hint: The function should take list data type as argument, not multiple integers)

def calculate_sum(list):
  sum = 0
  while(len(list)):
    sum = sum + list.pop()
  return sum

listc = [1,2,3,4,5]
sumOfList = calculate_sum(listc)
print(sumOfList)
```

```python
"""Task 6: Extract Prime Numbers"""
#Write a Python function named prime_numbers_in_list that takes a
#list of numbers as input and returns a new list containing only the prime
#numbers from the original list.
#(Hint: A prime number is a positive integer with exactly two divisors: 1 and itself)

def is_Prime(num):
  if num <= 1:
   return False
  for i in range(2,int(num**0.5)+1):
    if(num % i == 0):
      return False
  return True

def  prime_numbers_in_list(list):
    prime_numbers = []
    for n in list:
        if is_Prime(n):
            prime_numbers.append(n)
    return prime_numbers

# Example usage
my_list = [3, 4, 7, 10, 13, 15, 17, 20]
prime_list = prime_numbers_in_list(my_list)

print(f"Prime numbers in the list: {prime_list}")
```
