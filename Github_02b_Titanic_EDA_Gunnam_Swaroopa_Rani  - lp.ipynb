```python
"""Assignment-02b: Exploratory Data Analysis with Pandas
Context:
You are a data analyst tasked with performing exploratory data analysis (EDA)
on the Titanic dataset. This dataset contains information about passengers
aboard the Titanic, including demographics and survival status.
Your goal is to analyze the dataset using Pandas to gain insights into the
passengers and their characteristics."""
#Task 1: Loading the Data:
# Load the Titanic dataset into a Pandas DataFrame.
import pandas as pd

auto = pd.read_csv(r"C:\Users\Admin\Desktop\Swaroopa\Lithan_Phyton\el02_fc02_Assignment02\titanic.csv")
auto.head()
```

```python
"""Task 2: Initial Exploration:"""
#Display the first 5 rows of the dataset to get an overview.
auto.head(6)
```

```python
# Provide basic information about the dataset using the info() method.
auto.info()
```

```python
#Task 3: Handling Missing Values:
#Identify and count the number of missing values in each column.
auto.isnull().sum()
```

```python
# Decide on an appropriate strategy for handling missing values in at least
#two columns and implement it.

auto['Embarked'] = auto['Embarked'].fillna(auto['Embarked'].mode()[0])
auto['Cabin'] = auto['Cabin'].fillna(auto['Cabin'].mode()[0])
```

```python
# Verify that missing values have been handled
print("\nMissing values after handling:")
print(auto.isnull().sum())
```

```python
"""Task 4: Descriptive Statistics:"""
#Generate descriptive statistics for the numerical columns using the describe() method.
auto.describe()
```

```python
"""Task 5: Unique Values:"""
#List unique values in the 'Pclass' column.
auto['Pclass'].unique()
```

```python
#Count the occurrences of each unique value in the 'Embarked' column.
auto['Embarked'].value_counts()
```

```python
"""Task 6: Data Manipulation:"""
#Create a new column named 'FamilySize' that represents the sum of 'SibSp' and 'Parch'.
FamilySize = auto['SibSp'] + auto['Parch']
auto['FamilySize'] = FamilySize
auto.head()

```

```python
#Drop the 'Cabin' column from the dataset.
auto.drop('Cabin', axis=1, inplace=True)
auto.head()
```

```python
"""Task 7: Filtering and Sorting:"""
# Create a new DataFrame containing only passengers with an age greater than 30.
newage30 = auto[auto['Age']>30]
newage30.head(6)
```

```python
#Sort the DataFrame by 'Fare' in descending order.
auto.sort_values(by='Fare', ascending=False)
```

```python
"""Task 8: Data Analysis:"""
#Calculate the survival rate for different passenger classes ('Pclass').
auto.groupby('Pclass')['Survived'].mean()

```

```python
#Identify and display the passenger with the highest 'Fare'
auto[auto['Fare'] == auto['Fare'].max()]
```
